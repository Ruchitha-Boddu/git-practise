```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 40]

plt.plot(x, y)
plt.plot(x, y, marker='o')
plt.title("Sales growth")
plt.xlabel("Month")
plt.ylabel("Revenue")

```

```python
data= 10,10.5,10.510, 10.8,20,20.5,30
sales = 10,10.2,10.5
plt.hist(data, bins=10)
plt.hist(sales, bins=20)

```

```python
#bar chart 
categories = ['Electronics', 'Clothing', 'Furniture']
revenue = [50000, 30000, 40000]
plt.bar(categories, revenue)
plt.title("Revenue by Category")
plt.xlabel("Category")
plt.ylabel("Revenue")
plt.show()

```

```python
plt.bar(categories, revenue)

for i, v in enumerate(revenue):
    plt.text(i, v, f"${v}", ha='center', va='bottom')

plt.title("Revenue by Category")
plt.show()
```

```python
#scaler plot
ad_spend = [1000, 2000, 3000, 4000, 5000]
sales = [10, 25, 30, 45, 60]
plt.scatter(ad_spend, sales)
plt.title("Ad Spend vs Sales")
plt.xlabel("Ad Spend")
plt.ylabel("Sales")
plt.show()
```

```python
#histogram
import numpy as np

customer_age =  np.random.randint(18, 60, 100)
plt.hist(customer_age, bins=10)
plt.title("Customer Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
#multiple lines

days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
sales = [120, 150, 90, 180, 200]

sales_2024 = [100, 140, 160, 180, 220]
sales_2025 = [120, 150, 180, 210, 250]

plt.plot(days, sales_2024, label="2024")
plt.plot(days, sales_2025, label="2025")
plt.legend()
plt.title("Yearly Sales Comparison")
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12,5))

axes[0].bar(categories, revenue)
axes[0].set_title("Revenue")

axes[1].plot(days, sales)
axes[1].set_title("Sales Trend")

plt.show()
```

```python
plt.style.use("seaborn-v0_8")
plt.plot(days, sales, marker='o')
plt.title("Styled Sales Chart")
plt.show()
```

```python
#real dataframe plot
import pandas as pd
df = pd.DataFrame({
    "Category": categories,
    "Revenue": revenue
})

plt.bar(df["Category"], df["Revenue"])
plt.show()
```
